```python
import pandas as pd
```

```python
banking_parquet = 'data/banking.parquet'
sales_parquet = 'data/sales.parquet'
mcc_parquet = 'data/mcc.parquet'
```

```python
df_banking = pd.read_parquet(banking_parquet, engine='pyarrow')
```

```python
df_banking.head()
```

```python
df_mcc = pd.read_parquet(mcc_parquet, engine='pyarrow')
```

```python
df_mcc.head()
```

```python
comercial_descriptions = df_mcc['irs_description'].unique()
comercial_descriptions
```

```python
df_sales = pd.read_parquet(sales_parquet, engine='pyarrow')
```

```python
df_sales.head()
```

```python
sales_doc_id_n_state = df_sales[['document_id', 'state', 'mcc']]
```

```python
sales_doc_id_n_state.head()
```

```python
# removendo duplicados
unique_sales_doc_id_n_state = sales_doc_id_n_state.drop_duplicates(subset='document_id')
```

```python
unique_sales_doc_id_n_state.head()
```

```python
banking_by_state_n_mcc = pd.merge(df_banking, unique_sales_doc_id_n_state, on='document_id', how='left')
```

```python
banking_by_state_n_mcc.head()
```

```python
mcc_irs_description = df_mcc[['mcc', 'irs_description']]
```

```python
banking_by_state_n_mcc_descript = pd.merge(banking_by_state_n_mcc, mcc_irs_description, on='mcc', how='left')
```

```python
banking_by_state_n_mcc_descript.head()
```

```python
df = pd.DataFrame(banking_by_state_n_mcc_descript)

# Filtrar por tipo de transação 'pix_in'
df_pix_in = df[df['type'] == 'pix_in'].copy()
```

```python
# Transformar a coluna date_time em date e outra de hora
df_pix_in['date_time'] = pd.to_datetime(df_pix_in['date_time'])
df_pix_in['day'] = df_pix_in['date_time'].dt.day
df_pix_in['month'] = df_pix_in['date_time'].dt.month
df_pix_in['hour'] = df_pix_in['date_time'].dt.hour
df_pix_in['day_of_week'] = df_pix_in['date_time'].dt.day_name()
df_pix_in['time_of_day'] = df_pix_in['date_time'].dt.hour.apply(lambda x: 'Morning' if 5 <= x < 12 else 'Afternoon' if 12 <= x < 17 else 'Evening' if 17 <= x < 21 else 'Night')
df_pix_in.drop(columns=['date_time'])
```

```python
df_pix_in.to_csv('pix_in_banking.csv', index=False)
```
