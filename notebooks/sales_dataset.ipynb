```python
import pandas as pd
```

```python
banking_parquet = 'data/banking.parquet'
sales_parquet = 'data/sales.parquet'
mcc_parquet = 'data/mcc.parquet'
```

```python
df_banking = pd.read_parquet(banking_parquet, engine='pyarrow')
```

```python
df_banking.head()
```

```python
df_mcc = pd.read_parquet(mcc_parquet, engine='pyarrow')
```

```python
df_mcc.head()
```

```python
# Visualização do dataset de vendas por document_id, estado e data/hora
df_sales = pd.read_parquet(sales_parquet, engine='pyarrow')
```

```python
df_sales.head(30)
```

```python
# Agrupamento de todas as vendas por document_id
sales_per_doc_id = df_sales.groupby('document_id')
```

```python
sales_per_doc_id.head()
```

```python
# Carregamento do dataset de MCC / Descrição de estabelecimentos com base no MCC
mcc_description = df_mcc[['mcc', 'irs_description']]
```

```python
# Criação de dataset de vendas por tipos de estabelecimentos
sales_per_mcc_description = pd.merge(df_sales, mcc_description, on='mcc', how='inner')
```

```python
sales_per_mcc_description.head(20)
```

```python
# Agrupamento de descrição de estabelecimentos
irs_descriptions = sales_per_mcc_description.groupby('irs_description')
```

```python
irs_descriptions.head()
```

```python
# Limpeza de dados não importantes para aprendizado de maquina
sales_per_mcc = sales_per_mcc_description.drop(columns=['card_number', 'type'])
```

```python
sales_per_mcc.head()
```

### Tratamento dos dados

```python
# Tratamento da data/hora para novas colunas: Dia, Mes, Ano, Estações do ano
sales_per_mcc['date_time'] = pd.to_datetime(sales_per_mcc['date_time'])
sales_per_mcc['day'] = sales_per_mcc['date_time'].dt.day
sales_per_mcc['date'] = sales_per_mcc['date_time'].dt.date
sales_per_mcc['hour'] = sales_per_mcc['date_time'].dt.hour
sales_per_mcc['day_of_week'] = sales_per_mcc['date_time'].dt.day_name()
sales_per_mcc['month'] = sales_per_mcc['date_time'].dt.month
sales_per_mcc['year'] = sales_per_mcc['date_time'].dt.year
sales_per_mcc['time_of_day'] = sales_per_mcc['date_time'].dt.hour.apply(lambda x: 'Morning' if 5 <= x < 12 else 'Afternoon' if 12 <= x < 17 else 'Evening' if 17 <= x < 21 else 'Night')

```

```python
sales_per_mcc.head()
```

```python
sales_per_mcc.to_csv('sales_per_mcc.csv', index=False)
```
